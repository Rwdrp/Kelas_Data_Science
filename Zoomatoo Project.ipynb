```python
# Mengganti kode negara dengan nama negara
typo_kode = {
    '162': 'Philipines',
    '30' :'Brazil',
    '216':'United States',
    '14' :'Australia',
    '37' :'Canada',
    '184':"singapore",
    '214':"UAE",
    '1'  :'India',
    '94' :'Indonesia',
    '148':'New Zealand',
    '215':'United Kingdom',
    '166':'Qatar',
    '189':'South Africa',
    '191':'Sri Langka',
    '208':'Turkey'
}
dzr = dz['Country Code'] = dz['Country Code'].replace(typo_kode)
dzr
```

```python
dza = dz.rename(columns={'Country Code': 'Country'})
dza
```

```python
# Menghitung jumlah kemunculan Country dalam kolom
counts_country = dza['Country'].value_counts().reset_index()
counts_country.columns = ['Country', 'Count']
counts_country
```

```python
# Membuat diagram batang dari data tersebut
import matplotlib.pyplot as plt
plt.bar(counts_country['Country'], counts_country['Count'])
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Bar Chart - Country Counts')
plt.xticks(rotation=90)
plt.show()
```

```python
# Membuat diagram pie dari data tersebut
plt.figure(figsize=(8, 9))
plt.pie(counts_country['Count'], labels=counts_country['Country'], autopct='%1.1f%%', labeldistance=1.5)
plt.title('Pie Chart - Country Counts')
plt.axis('equal')
plt.show()
```

```python
print('Aggregate rating mean:', dza['Aggregate rating'].mean())
```

```python
# Menghitung rata-rata aggregate rating dari setiap negara
dza_group = dza.groupby('Country')['Aggregate rating'].mean()
dza_group
```

```python
# Membuat diagram batang dari data tersebut
plt.bar(dza_group.index, dza_group.values)
plt.xlabel('Country')
plt.ylabel('Aggregate rating')
plt.title('Bar Chart - Country Aggregate rating')
plt.xticks(rotation=90)
plt.show()
```

```python
dza_grouped = dza.groupby('Country')['Rating text'].value_counts()
dza_grouped = dza_grouped.rename('Count').reset_index()
dza_grouped
```

```python
# Membuat diagram batang dari data tersebut
stacked_data = dza_grouped.pivot(index='Country', columns='Rating text', values='Count')
stacked_data.plot(kind='bar',stacked=True)
plt.xlabel('Country')
plt.ylabel('Rating textg')
plt.title('Bar Chart - Country Rating text')
plt.xticks(rotation=90)
plt.show()
```

```python
data_ind = dza[dza['Country'] == 'India']
data_ind
```

```python
# Membuat diagram batang dari data tersebut
dataind_grouped = data_ind.groupby('Country')['Rating text'].value_counts()
dataind_grouped = dataind_grouped.rename('Count').reset_index()
dataind_grouped
india_data = dataind_grouped.pivot(index='Country', columns='Rating text', values='Count')
india_data.plot(kind='bar',stacked=True)
plt.xlabel('Country')
plt.ylabel('Rating textg')
plt.title('India Bar Chart - Country Rating text')
plt.xticks(rotation=90)
plt.show()
```
