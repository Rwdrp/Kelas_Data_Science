## 2. Mana yang lebih baik? Penjualan di Jerman atau Inggris?

Anda diminta untuk menganalisa performa penjualan di Jerman dan Inggris. Penjualan yang dimaksud adalah pendapatan bulanan yang berasal dari kedua negara tersebut. Tujuan analisis ini ingin menentukan lokasi cabang baru.

```python
data['Bulan'] = data['InvoiceDate'].dt.month
data['Tahun'] = data['InvoiceDate'].dt.year
data['Revenue'] = data['Quantity']*data['UnitPrice']
data
```

```python
data_update = data[data['Quantity'] > 0]
data_update
```

```python
UK = data_update[data_update['Country']=='United Kingdom'][['Tahun','Bulan','Revenue']].groupby(['Tahun','Bulan']).sum()
DE = data_update[data_update['Country']=='Germany'][['Tahun','Bulan','Revenue']].groupby(['Tahun','Bulan']).sum()
print(UK), print(DE)
```

```python
print('Rata-rata pendapatan bulanan UK =', UK['Revenue'].mean())
print('Rata-rata pendapatan bulanan Germany =', DE['Revenue'].mean())
```

Penjualan rata-rata di UK jauh lebih tinggi dibandingkan Germany. Artinya performa penjualan di UK lebih baik pada satu tahun ke belakang.

Melakukan Uji Hipotesis Two Sample Independent
==> karena ingin menguji perbedaan signifikan "data populasi" antara penjualan di UK dan Germany dari data sampel setahun kebelakang

H0: Rata-rata pendapatan bulanan UK <= Germany

H1: Rata-rata pendapatan bulanan UK > Germany 

```python
t_stats, p_val=stats.ttest_ind(UK['Revenue'],DE['Revenue'])
print('T-Statistic:', t_stats)
print('P-value:', p_val/2) # ==> dibagi 2 karena one sided
```

critical value = 0.05

p-value = 2.33 x 10^-9

critical value < p-value


\#Kesimpulan
Dari pengujian statistik didapatkan bahwasanya penjulan di UK akan lebih well-performed dibandingkan di Germany. Maka yang paling bagus untuk membuka cabang di UK untuk menggait pembeli yang berasal dari Geramany meilhat posisi geografis juga lebih dekat.

## 1. Apakah Ada Hubungan Antara Gender dengan Kebiasaan Merokok?

Melakukan uji hipotesis dengan tipe chi-square

H0: Tidak ada hubungan antara Gender dengan Kebiasaan Merokok (Pria dan Wanita sama)

H1 : Ada hubungan antara Gender dengan Kebiasaan Merokok (Pria dan Wanita berbeda)

```python
# Membuat tabel kontigensi / contingency table
contingency_table = pd.crosstab(ins['sex'], ins['smoker'])
contingency_table
```

```python
stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)
print("P-value:",p_val)
```

Critical value: 0.05

p-value = 0.005 < critical value

H0 ditolak

**Kesimpulan**: 


Ada hubungan antara jenis kelamin dengan kebiasaan merokok, dimana pria yang merokok lebih besar dibandingkan wanita.

## 2. Apakah Biaya Premi Laki-laki dan Perempuan Sama?

```python
pria = ins[ins['sex']=='male']['charges']
wanita = ins[ins['sex']=='female']['charges']

print("Rata-rata premi pria:", pria.mean())
print("Rata-rata premi wanita:", wanita.mean())
```

Melakukan Uji Hipotesis Two Sample Independent Two Sided

H0: Rata-rata premi pria = premi wanita

H1: Rata-rata premi pria != premi wanita

```python
t_stat, p_val = stats.ttest_ind(pria,wanita)
print("P-value:", p_val)
```

Manakah jenis tempat penampungan yang sebaiknya dibuat? Jenis garbage collection atau organics?

```python
ds['Bulan'] = ds['report_date'].dt.month
ds
```

```python
#Total jenis sampah 
gar_coll = ds[ds['load_type'] == 'GARBAGE COLLECTIONS'][['Bulan', 'load_weight']].groupby('Bulan').sum()
organics = ds[ds['load_type'] == 'ORGANICS'][['Bulan', 'load_weight']].groupby('Bulan').sum()
```

```python
print('Rata-rata garbage collection =', gar_coll['load_weight'].mean())
print('Rata-rata organics =', organics['load_weight'].mean())
```

Bahwa garbage collection selama 7 bulan terakhir lebih banyak dibandingkan dengan sampah organics.

Melakukan Uji hipotesis Two Sample Independent
--> karena ingin menguji perbedaan signifikan "data populasi" antara garbage collection dengan organics dari data sampel 7 bulan kebelakang.

H0 : Rata-rata garbage collection <= organics

H1 : Rata-rata garbage collection >  organics

```python
t_stat, p_val = stats.ttest_ind(gar_coll['load_weight'], organics['load_weight'])
print('T-Statistic:', t_stat)
print('P-value:', p_val/2) # --> dibagi 2 karena one sided
```

critical value = 0.05

p-value = 0.005 < critical value

Hasil : H0 ditolak

Dari pengujian secara statsistik diperoleh bahwa jumlah sampah garbage collection dan organics sangat berbeda. Sampah garbage collection akan lebih banyak ditampung dibandingkan sampah irganics. Sehingga akan lebih bagus apabila pembuatan tempat penampungan yang baru yakni jenis garbage collection melihat trend yang cukup besar.
